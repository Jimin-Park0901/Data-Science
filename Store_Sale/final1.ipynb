```python
%matplotlib inline

import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('store_sales.csv')
display(df)
```

# Analysis #1

```python
sns.set_theme(style='ticks')
plt.figure(figsize=(5, 5))
sns.histplot(data=df, x='competing_stores ')

plt.show()

display(df.describe())
```

```python
sns.countplot(data=df, x='parking', palette=['#5176a8', '#d8855b'])
plt.show()
```

```python
sns.countplot(data=df, x='dining', palette=['#5176a8', '#d8855b'])
plt.show()
```

```python
sns.countplot(data=df, x='main_street', palette=['#5176a8', '#d8855b'])
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=df, x='area ', y='sales')
plt.show()
correlation = df['area '].corr(df['sales'])
print(correlation)
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=df, x='walking_time', y='sales')
plt.show()
correlation = df['walking_time'].corr(df['sales'])
print(correlation)
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=df, x='competing_stores ', y='sales')
plt.show()
correlation = df['competing_stores '].corr(df['sales'])
print(correlation)
```

```python
plt.figure(figsize=(5, 5))
sns.scatterplot(data=df, x='population_density', y='sales')
plt.show()
correlation = df['population_density'].corr(df['sales'])
print(correlation)
```

```python
plt.figure(figsize=(5, 5))
sns.histplot(data=df, x='sales', hue="parking")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.histplot(data=df, x='sales', hue="dining")
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.histplot(data=df, x='sales', hue="main_street")
plt.show()
```
